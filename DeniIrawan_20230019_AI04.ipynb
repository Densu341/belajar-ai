# Regresi Linier

Regresi Linier (Linear Regression/LR) adalah algoritma pembelajaran mesin berdasarkan pembelajaran yang diawasi. LR melakukan tugas regresi. Model regresi memprediksi nilai target berdasarkan variabel independen. LR biasa digunakan untuk mengetahui hubungan antara variabel dan prediksi. Model regresi berbeda-beda berdasarkan jenis hubungan antara variabel dependen dan independennya, dengan mempertimbangkan jumlah variabel independen yang digunakan. Untuk mudahnya, LR bertugas membuat garis pemisah antar kelompok target.

Praktikum ini akan mendemonstrasikan cara menggunakan berbagai pustaka Python untuk mengimplementasikan regresi linier pada dataset diabetes. Kita akan mendemonstrasikan model linier biner karena ini akan lebih mudah untuk divisualisasikan.

Pada contoh di bawah ini, kita hanya menggunakan fitur pertama dari dataset diabetes, untuk mengilustrasikan titik data dalam plot dua dimensi. Garis lurus dapat dilihat pada plot, menunjukkan bagaimana regresi linier mencoba menggambar garis lurus yang paling baik, meminimalkan jumlah kuadrat sisa antara respons yang diamati dalam dataset, dan respons yang diprediksi oleh pendekatan linier.

image.png

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

## Memuat dataset diabetes

image.png

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

## Memilih satu fitur saja

image.png

```python
diabetes_fitur = diabetes_X[:, np.newaxis, 2]
print("Jumlah total data :", len(diabetes_fitur))
```

## Memecah data fitur menjadi data latih dan data uji

image.png

```python
diabetes_X_train = diabetes_fitur[:-20]
diabetes_X_test = diabetes_fitur[-20:]
print("Jumlah data latih :", len(diabetes_X_train))
print("Jumlah data uji :", len(diabetes_X_test))
```

## Memecah data target menjadi data latih dan data uji

image.png

```python
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
print("Jumlah data latih :", len(diabetes_y_train))
print("Jumlah data uji :", len(diabetes_y_test))
```

## Membuat model Regresi Linier

image.png

```python
regresi = linear_model.LinearRegression()
```

## Melatih model menggunakan data latih

image.png

```python
regresi.fit(diabetes_X_train, diabetes_y_train)
```

## Membuat prediksi menggunakan data uji

image.png

```python
diabetes_y_pred = regresi.predict(diabetes_X_test)
```

## Menghitung koefisien regresi

image.png

```python
print("Coefficients :", regresi.coef_)
```

## Menghitung Mean Squared Error (MSE)

image.png

```python
print("Mean Squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
```

## Menghitung koefisien determinasi

Prediksi semakin bagus bila koefisien determinasi mendekati atau sama dengan 1

image.png

```python
print("Koefisien Determinasi : %.2f" %r2_score(diabetes_y_test, diabetes_y_pred))
```

## Visualisasi Output

image.png

```python
plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
